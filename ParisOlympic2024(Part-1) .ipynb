```python
data.drop(columns=['name','name_short','country','birth_place','residence_place','nickname','hobbies','family','reason','hero','influence','philosophy','sporting_relatives','ritual','other_sports'],inplace= True)
```

```python
data.drop(columns=['height','weight'],inplace=True)
data.drop(columns='nationality',inplace=True)
data.drop(columns=['coach'],inplace=True)
data.drop(columns=['birth_country'],inplace=True)
```

```python
data.head()
```

```python
data.birth_country = data.birth_country.fillna(data.residence_country)
data.residence_country = data.residence_country.fillna(data.birth_country)
```

```python
data.rename(columns={'name_tv	':'name,','country_full':'country','nationality_full':'nationality'},inplace=True)
```

```python
gender_classification = dict(data.gender.value_counts())
```

```python
gender_classification
```

```python
import matplotlib.pyplot as plt
plt.pie(gender_classification.values(),labels=gender_classification.keys(),shadow=True,autopct="%0.1f%%")
```

```python
# Top countries
top_10_participated_country=pd.DataFrame()
top_10_participated_country['countries'] = data.country.drop_duplicates()
l=[]
for i in top_10_participated_country.countries:
  l.append(data[data.country==i].country.count())
top_10_participated_country['No_Of_Athletics'] = l
top_10_participated_country = top_10_participated_country.sort_values(by=['No_Of_Athletics'], ascending=True)
top_10_participated_country=top_10_participated_country.iloc[-10:,:]
```

```python
top_10_participated_country
```

```python
plt.barh(top_10_participated_country.countries,top_10_participated_country.No_Of_Athletics)
plt.title('Top 10 countries participated in paris olympic')
plt.xlabel('Contries')
plt.ylabel('No of Athletics')
plt.show()
```
