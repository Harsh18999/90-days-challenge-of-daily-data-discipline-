# **Implementation of Perceptrone Trick**

```python
# create a classification dataset
from sklearn.datasets import make_classification
import numpy as np

X,Y = make_classification(n_samples=100,n_features=2,n_classes=2,n_informative=1,n_redundant=0,n_clusters_per_class=1,hypercube=False,random_state=41,class_sep=10)
```

```python
X
```

```python
# plot of dataset
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.scatter(X[:,0],X[:,1],c=Y, cmap='winter',s=100)
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
# define perceptrone function
def perceptrone(X,Y):
  X = np.insert(X,0,1,axis=1)
  weight = np.ones(X.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)
    y_pred = step(np.dot(weight,X[j]))
    weight += (Y[j] - y_pred)*lr*X[j]

  return weight[0],weight[1:]
```

```python
# find intercept and coefficiants of data
intercept_,coef = perceptrone(X,Y)
```

```python
print(intercept_)
print(coef)
```

```python
m = -(coef[0]/coef[1])
b = -(intercept_/coef[1])
```

```python
print(min(X[:,0]))
print(max(X[:,0]))
print(min(X[:,1]))
print(max(X[:,1]))
```

```python
x_input = np.linspace(-3,2.15,100)
y_input = m*x_input + b
```

```python
# plot overall output
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=Y,cmap='winter',s=100)
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.ylim(-3,2)
```
